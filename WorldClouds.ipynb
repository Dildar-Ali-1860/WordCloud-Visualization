```python
from wordcloud import WordCloud, STOPWORDS     
```

```python
import pandas as pd
import numpy as np
```

```python
import urllib
data = urllib.request.urlopen('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/alice_novel.txt').read().decode("utf-8")
data  
```

```python
stopwords = set(STOPWORDS)
stopwords  
```

```python
data_wc = WordCloud(background_color='white',max_words=2000,stopwords=stopwords)  
```

```python
data_wc.generate(data)
```

```python
data
```

```python
import matplotlib.pyplot as plt
plt.imshow(data_wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
fig=plt.figure(figsize=(15,8))    

plt.imshow(data_wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
stopwords.add('said') 
data_wc.generate(data)
fig=plt.figure(figsize=(16,8))   
plt.imshow(data_wc)
```

```python
# save mask to alice_mask
from PIL import Image
data_mask = np.array(Image.open(urllib.request.urlopen('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/labs/Module%204/images/alice_mask.png')))
```

```python
data_mask
```

```python
fig = plt.figure(figsize=(16,8))

plt.imshow(data_mask,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# instantiate a word cloud object
data_wc = WordCloud(background_color='white', max_words=2000, mask=data_mask, stopwords=stopwords)

# generate the word cloud
data_wc.generate(data)

# display the word cloud
fig = plt.figure(figsize=(14, 18))

plt.imshow(data_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
